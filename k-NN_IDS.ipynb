```python
nfeatures = train_df.columns.tolist()
#Remove categorical data
nfeatures.remove('label')
nfeatures.remove('protocol_type')
nfeatures.remove('service')
nfeatures.remove('flag')

#Create dataframe with only numerical features as float type
data_features = train_df[nfeatures].astype(float)
```

```python
#Change target labesl to binary labels - normal or malicious
binary_labels = train_df['label'].copy()
binary_labels[binary_labels != 'normal'] = 'malicious'
binary_labels.value_counts()
binary_labels
```

```python
len(train_df['label'].unique())
#Convert all non-normal labels to malicious
train_df['label'][train_df['label'] != 'normal'] = 'malicious'
```

```python
import sklearn.naive_bayes 
from time import time

#Training a classifier
classifier = sklearn.naive_bayes.GaussianNB()
time_start = time()
classifier.fit(data_features, binary_labels)
time_present = time() - time_start
print ("Classifier trained in %f seconds." % round(time_present, 3))

```

```python
import sklearn.model_selection as sk

X_train, X_test, y_train, y_test = sk.train_test_split(train_df[nfeatures], train_df['label'], test_size=0.2,random_state=42)
```

```python
#Predictions on the test data
classifier.fit(data_features, binary_labels)
t0 = time()
pred = classifier.predict(X_test)
tt = time() - t0
print ("Predicted in {} seconds".format(round(tt,3)))

```

```python
#Calculating out the accuracy
from sklearn.metrics import accuracy_score
acc = accuracy_score(pred, y_test)
print ("Accuracy is {}.".format(round(acc,4)))
```
